```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('rolling_stock_data.csv')
```

```python
X = data[['Temperature', 'Vibration']]
y = data['Label']
```

```python
model = IsolationForest(random_state=0)
model.fit(X)

y_pred = model.predict(X)
y_pred = np.where(y_pred == -1, 1, 0)  # Convert -1 to 1 for anomalies
```

```python
accuracy = np.mean(y_pred == y)
print("Accuracy:", accuracy)
```

```python
plt.scatter(X['Temperature'], X['Vibration'], c=y_pred, cmap='coolwarm')
plt.xlabel('Temperature')
plt.ylabel('Vibration')
plt.title('Anomaly Detection in Rolling Stock Data')
plt.colorbar(label='Prediction (0: Normal, 1: Anomaly)')
plt.show()
```
